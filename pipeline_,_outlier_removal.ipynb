```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/covid_toy.csv")
df.head(2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder , StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns = ["has_covid"])
y = df["has_covid"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
# define the columns that need to be processed
categorical_feature = ["gender" , "city"]
numerical_feature = ["age","fever"]
```

```python
# create transformer
numeric_transformer = Pipeline(steps = [
    ("imputer",SimpleImputer(strategy = "mean")),
    ("scaler",StandardScaler())
])
```

```python
categorical_transformer = Pipeline(steps = [("imputer",SimpleImputer(strategy = "most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown = "ignore"))])
```

```python
# combine transformers
preprocessor = ColumnTransformer(transformers = [
    ("num",numeric_transformer,numerical_feature),
    ("cat",categorical_transformer,categorical_feature)
])
```

```python
preprocessor.fit_transform(x_train)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/newplacementdata.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df["placement_exam_marks"].describe()
```

```python
sns.boxplot(x = df["placement_exam_marks"] )
```

```python
# finding the irq
percentile25 =  df["placement_exam_marks"].quantile(0.25)
percentile75 =  df["placement_exam_marks"].quantile(0.75)
```

```python
percentile25
```

```python
percentile75
```

```python
IQR =percentile75-percentile25
IQR
```

```python
upper_limit = percentile75 + 1.5*IQR
upper_limit
```

```python
lower_limit = percentile25 - 1.5*IQR
lower_limit
```

```python
# Finding out outliers
```

```python
df[df["placement_exam_marks"] > upper_limit]
```

```python
df[df["placement_exam_marks"] < lower_limit]
```

```python
# trimming ---> 4(lower) ,,,1 , upper(25)   ....(50) ---> here 1,50 are outliers
```

```python
# trimming (outlier.removing technique 1)
```

```python
newdf = df[df["placement_exam_marks"]< upper_limit]
```

```python
# newdf
```

```python
# comparison
```

```python
plt.figure(figsize = (15,5))


plt.figure(222)
sns.boxplot(x = df["placement_exam_marks"])
plt.figure(224)
sns.boxplot(x = newdf["placement_exam_marks"])
plt.show()
```

```python
# capping (outlier.removing technique 2 )
```

```python
newdf_cap = df.copy()
```

```python
# min = 5 , max = 15
# min 4,3,1
# max = 20,30,50
# updated_min_value = 1

# updated_max_value = 50
```

```python
newdf_cap["placement_exam_marks"] = np.where(

                                             newdf_cap["placement_exam_marks"] > upper_limit,
                                                 upper_limit,

                                             np.where(
                                                      newdf_cap["placement_exam_marks"] < lower_limit,
                                                      lower_limit,
                                                      newdf_cap["placement_exam_marks"])

                                             )
```

```python
newdf_cap
```

```python
newdf_cap.shape
```

```python
# comparison
```

```python
plt.figure(figsize = (15,8))


plt.figure(222)
sns.boxplot(x = df["placement_exam_marks"])
plt.figure(224)
sns.boxplot(x = newdf_cap["placement_exam_marks"])
plt.show()
```

```python

# cca ---> complete case analysis
# df = missing data ===> filter missing data--->
# new_df = filtered_columns in which we have missing values.

# new_df -----> missing_data_drop


# architecture ----> histogram--->
# past column in which we have missing values.

# if past data achitecture is overlap to new data architecture --->
# it means we can drop missing data
# if past data achitecture is not overlap to new data architecture --->
# it means we can not drop missing data
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/dsjob.csv")
df.head()
```

```python
df.isnull().mean()*100
```

```python
cols = [var for var in df.columns if df[var].isnull().mean()<0.05 and df[var].isnull().mean()>0]
```

```python
cols
```

```python
df[cols].sample(5)
```

```python
df["education_level"].value_counts()
```

```python
len(df[cols].dropna())/ len(df)                        #we will g=have 94% of data remaining after removing missing values
```

```python
new_df = df[cols].dropna()
df.shape , new_df.shape
```

```python
import matplotlib.pyplot as plt

fig  = plt.figure()
ax = fig.add_subplot(111)

# origginaldata

df["experience"].hist(bins = 50 , ax = ax ,density = True , color = "red")

# data after cca , the argument alpha_makes  the color transparent , so we can
# see the overlap of the 2 distribution

new_df["experience"].hist(bins = 50 , ax = ax ,density = True , color = "green" , alpha = 0.8)

```

```python
temp = pd.concat([
                #percentage of observations per category , original data
                df["enrolled_university"].value_counts()/ len(df) ,
                # percentage of observations per category , cca data
                new_df["enrolled_university"].value_counts()/ len(new_df)
                ],
                 axis = 1)
# add the column name
temp.columns = ["original" , "cca"]
temp
```
