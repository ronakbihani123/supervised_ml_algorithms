```python
import pandas as pd
import numpy as np
```

```python
 df = pd.read_csv("/content/insurance.csv")
df.head(3)
```

```python
# df.isnull().sum()
```

```python
df = pd.get_dummies(df,columns=["sex","smoker","region"],drop_first=True)
df.head(3)
```

```python
df = df.astype(int)
```

```python
df.head(3)
```

```python
x = df.drop(columns = ["charges"])
y = df["charges"]
#
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred  = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
df =pd.read_csv("/content/BMW sales data (2010-2024).csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = pd.get_dummies(df,columns=["Model","Region","Color" , "Fuel_Type" , "Transmission" , "Sales_Classification"],drop_first=True)
df.head(3)
```

```python
df = df.astype(int)
```

```python
df.head(3)
```

```python
x = df.drop(columns = ["Sales_Volume"])
y = df["Sales_Volume"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred  = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
