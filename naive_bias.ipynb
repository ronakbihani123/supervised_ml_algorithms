```python
# what is naive bayes?
# naive bayes is a probabilistic machine learning model based on bayes theorem . it is mostly used for classification problem such as email spam detection , medical diagnosis , and
# it is called naive because it assumes all features are independent to each other and it is called bias because it is based on bayes theorem of prabability .

# bayes theorem is used to update the probability of a hypothesis class based on new evidence.
# formula = probability of c/x = probability(x/c)*probability(c)/ probability(x)
# where probability(c/x) ==> it is probability of class c given feature x
# probabiility of x/c ===> probability of feature x in a given class c.
# probabiility of c ===> prior probability of class
# probabiility of x====> evidence = probability of feature of all classes .


```

```python
# advantage of naive bias >
# it is very fast and efficient .
# it works well with high dimensional data .
# it requires small training data .
# it handles noisy data .
```

```python
# limitations
# it assumes of independence of features
# it is not suitable for features with strong correlation
```

```python
# task ---> suppose we want to classify whether a person buy a computer (yes/no) based on two features.
# 1. age  ----> young , middle , old
# 2.income ---> high ,medium , low
```

```python
# person              age                   income                   buys_computer
# first               young                 high                     no
# second              young                 medium                   no
# third               middle                high                     yes
# fourth              old                   medium                   yes
# fifth               old                   low                      yes
# sixth               middle                low                      no
```

```python
# step1 ---> we calculate how many people bought vs how many people did not buy a computer
# total  = 6 ,  buy = 3(yes)  ,  not buy = 3(no)
# probability of yes = 0.5 , probability of no = 0.5

```

```python
# step2 ---> now we compute probabilities of features given the class .
# for class = yes
# age(middle) = 1/3
# age(old) = 2/3
# income(high) = 1/3
# income(medium) = 1/3
# income(low) = 1/3
# for class  = no
# age(young) = 2/3
# age(middle) = 1/3
# age(old) = 0
# income(high) = 1/3
# income(medium) = 1/3
# income(low) = 1/3
```

```python
# step 3 ---> now suppose we want to predict for a new person whose age = old and income = medium
# formula = prob(yes)/old and medium = prob(yes)*prob(old/yes)*prob(medium/yes)
        # = 0.5*2/3*1/3
        # = 0.11
  # formula = prob(no)/old and medium = prob(no)*prob(old/no)*prob(medium/no)
                                    = 0.5 * 0 * 1/3
                                    = 0
```

```python
# step 4 ----> decision
# probability of yes/ old and medium = 0.111
# probability of no/ old and medium = 0
# model will predict whose probabilty is more .. the probability of yes will be more so model will predict yes.
```

```python
# types of naive bias
# 1 ) multinomial naive bias ===> for text classification and word counts.
# 2 ) bernoulli naive bias  ===> for binary features ex : word present or not.
# 3) gaussian naive bias ---> for continuous features or normal distribution.
```

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Social_Network_Ads (1).csv" , usecols = ["Age" , "EstimatedSalary" , "Purchased"])
```

```python
df.head(2)
```

```python
x = df.drop(columns = ["Purchased"] )     #independent variable
y = df["Purchased"]                       #dependent variable
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test  = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_new = sc.fit_transform(x_train)
```

```python
x_test_new = sc.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
classifier = GaussianNB()
```

```python
classifier.fit(x_train_new , y_train)
```

```python
y_pred= classifier.predict(x_test_new)
```

```python
y_pred
```

```python
from sklearn.metrics import recall_score , confusion_matrix , f1_score ,  precision_score
```

```python
confusion_matrix(y_test , y_pred)
```

```python
recall_score(y_test , y_pred)
```

```python
precision_score(y_test , y_pred)
```

```python
f1_score(y_test , y_pred)
```
