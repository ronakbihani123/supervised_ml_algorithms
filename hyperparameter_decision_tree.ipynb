```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv("/content/Social_Network_Ads (1).csv")
```

```python
def analyzer(max_depth):
  data = pd.read_csv("/content/Social_Network_Ads (1).csv")
  x = data.iloc[: , 2:4].values
  y = data.iloc[: , -1].values
  clf = DecisionTreeClassifier(max_depth=max_depth)
  clf.fit(x,y)
  a = np.arange(start=x[:,0].min()-1 , stop=x[:,0].max()+1 , step=0.01)                 #why these points are used ?
  b = np.arange(start=x[:,1].min()-1 , stop=x[:,1].max()+1 , step=1000)
  xx, yy = np.meshgrid(a,b)
  input_array = np.array([xx.ravel() , yy.ravel()]).T
  labels = clf.predict(input_array)
  labels = labels.reshape(xx.shape)
  plt.contourf(xx , yy , labels , alpha=0.5 )
  plt.scatter(x[:,0] , x[:,1] , c=y , alpha=0.5)
```

```python
analyzer(max_depth=5)
```
