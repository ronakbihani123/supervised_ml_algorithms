```python
# a random forest is an ensemble machine learning model that combines multiple decision trees to make predictions
```

```python
# it is mostly used for classificatiom (predicting categories) but can also be used for regression (predicting numbers) . it reduce the limitation of a single decision tree
# like overfitting by combining many trees.
```

```python
# how it works?
# 1). bootstrep sampling (bagging)----> the algorithm randomly select subset of data to train each tree .this means every tree gets slightly different data
# increasing diversity
# 2). random feature selection ---> at each split in a tree instead of considering all features it fix a random subset of feature . this ensures trees do not all look the same .
# 3). growing decision tree ---> each tree is trained independently using the sampled data and features . trees are usually grown deep so that they can capture patterns.
# 4). predicting  ---> for classification each tree predicts a class label and the majority vote decides the final prediction .
# 5). for regression -------> the everage of all trees predictions is taken .
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/Social_Network_Ads.csv")
df.head(2)
```

```python
df.isnull().sum()
```

```python
df["Gender"] = df["Gender"].map({"Male" : 1 , "Female" : 0})
```

```python
df.head()
```

```python
df = df.drop(columns = ["gender"])
```

```python
x = df.drop(columns = ["Purchased"])
y = df["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y ,test_size=0.2 , random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train , y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```

```python
# fine tuning
```

```python
from sklearn.model_selection import GridSearchCV , RandomizedSearchCV
```

```python
rf = RandomForestClassifier(
    n_estimators=80 ,
    max_features = "sqrt" ,
    max_depth=6,
    max_leaf_nodes=6
)
```

```python
rf.fit(x_train , y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
accuracy_score(y_test , y_pred)
```

```python
param_grid = {
    "n_estimators" : [50,100,200],
    "max_depth" : [None , 10,20,30],
    "min_samples_split" : [2,5,8],
    "min_samples_leaf" : [1,2,4],
    "bootstrap" : [True , False]
}
```

```python
grid_search =  GridSearchCV(RandomForestClassifier() , param_grid=param_grid , cv=5)
```

```python
grid_search.fit(x_train , y_train)
```

```python
print("Best Parameters" , grid_search.best_params_)
print("Best Estimator" , grid_search.best_estimator_)
```

```python
model = RandomForestClassifier(max_depth = 3 ,
                               max_features = "log2" ,
                               max_leaf_nodes = 3 ,
                               n_estimators = 50)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
accuracy_score(y_test , y_pred)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/insurance.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = pd.get_dummies(df , columns = ["sex" , "smoker" , "region"])
```

```python
df.head(2)
```

```python
df.astype(int)
```

```python
x = df.drop(columns = ["charges"])
y = df["charges"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y ,test_size=0.2 , random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
x_train_new = pd.DataFrame(x_train_sc , columns = x.columns)
```

```python
x_train_new
```

```python
x_test_new = pd.DataFrame(x_test_sc , columns = x.columns)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(x_train_new , y_train)
```

```python
y_pred = rf.predict(x_test_new)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test , y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "n_estimators" : [100,200],
    "max_depth" : [None , 10,20],
    "min_samples_split" : [2,5],
    "min_samples_leaf" : [1,2],
    "bootstrap": [True , False]
}
```

```python
grid_search = GridSearchCV(RandomForestRegressor() , param_grid=param_grid , cv=5)
grid_search.fit(x_train_new , y_train)

print("Best Parameters"  , grid_search.best_params_)
print("Best Estimator" , grid_search.best_estimator_)
```

```python
y_pred = grid_search.predict(x_test_new)
r2_score(y_test , y_pred)
```

```python
df = pd.read_csv("/content/titanic.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df["Fare"] = df["Fare"].fillna(df["Fare"].mean())
```

```python
df = df.drop(columns = ["Cabin" , "Name" , "Ticket"])
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df["Sex"] = lb.fit_transform(df["Sex"])
df["Embarked"] = lb.fit_transform(df["Embarked"])
```

```python
df.head(2)
```

```python
x = df.drop(columns = ["Survived"])
y = df["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y ,test_size=0.2 , random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```

```python
param_grid = {
    'n_estimators': [100,200,500],
    "max_depth" : [None , 10 , 20 , 30],
    "min_samples_split" : [5,10,15],
    "min_samples_leaf" : [1 ,2 , 4] ,
    "max_features" : ["sqrt" , "log2"]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(RandomForestClassifier() , param_grid=param_grid , cv=3)
grid_search.fit(x_train , y_train)

print("Best Parameters"  , grid_search.best_params_)
print("Best Estimator" , grid_search.best_estimator_)
```

```python
y_pred = grid_search.predict(x_test)
accuracy_score(y_test , y_pred)
```
