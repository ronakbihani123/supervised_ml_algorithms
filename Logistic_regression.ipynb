```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("covid_toy.csv")
df.head(2)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer()
df["fever"] = si.fit_transform(df[["fever"]])
```

```python
x = df.drop(columns = ["has_covid"])
y = df["has_covid"]
```

```python
x = pd.get_dummies(x , columns=["gender" , "cough" , "city"])
```

```python
x = x.astype(np.int32)
```

```python
x.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
y = lb.inverse_transform(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
x_train.head(2)
```

```python
y_train = pd.DataFrame(y_train , columns = ["has_covid"])
```

```python
y_train.head(2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train , y_train )
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
param_grid = [
    {"penalty": ["l1","l2" , 'elasticnet' , 'none'],
     "C" : np.logspace(-4,4,20),
     "solver" : ["lbfgs" , "newton-cg" , "liblinear" , "sag" , "saga"],
     "max_iter" : [10,50,100 , 1000]}
]
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(lr , param_grid = param_grid , cv = 6, verbose = True , n_jobs = -1)
clf
```

```python
best_clf = clf.fit(x_train,y_train)
best_clf.best_estimator_
```

```python
print(f"accuracy : {best_clf.score(x_train , y_train ):.3f}")
```
